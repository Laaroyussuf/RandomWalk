```python
from random import choice
import matplotlib.pyplot as plt
class RandomWalk:
    #A class to generate random walk
    def __init__(self, num_point = 5000):
        #initialized attribute of walk
        self.num_point=num_point
        self.x_value = [0]
        self.y_value = [0]
        
    def get_step(self):
        #This tells the number of steps.
        direction = choice([1, -1])
        distance = choice([0, 1, 2, 3, 4])
        return direction*distance
    
    
    def fill_walk(self):
        #calculate all points in walk
        while len(self.x_value) < self.num_point:
            #Keep taking steps until the walk reaches the desired length.
            x_step = self.get_step()
            y_step = self.get_step()
            if x_step == 0 and y_step == 0:
                #if there is no steps taken at a point,it continues with the next.
                continue
                
            x = self.x_value[-1] + x_step
            y = self.y_value[-1] + y_step
            
            self.x_value.append(x)
            self.y_value.append(y)
            
            
rw = RandomWalk()
rw.fill_walk()
plt.style.use('classic')
fig, ax = plt.subplots()
point_num = range(rw.num_point)
ax.scatter(rw.x_value,rw.y_value, c=point_num, cmap=plt.cm.Blues, edgecolors='none', s=15)
ax.scatter(0, 0, c='green', s=100)
#This indicates the starting point.
ax.scatter(rw.x_value[-1], rw.y_value[-1], c='red', s=100)
# This indicate the destination
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
#This indicate the last point i.e the destination
plt.show()
```
